```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
from plthelper import *

%config InlineBackend.print_figure_kwargs={'facecolor' : "w"}

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 12
```

```python
# Load data

filename = "sample-gamma.txt"
output_directory = "../files/input/"
path = output_directory + filename

data = pd.read_csv(path, sep = ", ", header = None, skiprows=23)
data.columns = ["Time [s]", "<E>_e [MeV]", "<E^2>_e [MeV]", "L_e [erg/s]", "<E>_eb [MeV]", "<E^2>_eb [MeV]", "L_eb [erg/s]", "<E>_x [MeV]", "<E^2>_x [MeV]", "L_x [erg/s]"]

data
```

```python
time = data.iloc[:,0].values * 1000 # in ms
luminosity = [data.iloc[:,3].values, data.iloc[:,6].values, data.iloc[:,9].values]
mean_energy = [data.iloc[:,1].values, data.iloc[:,4].values, data.iloc[:,7].values]
mean_squared_energy = lumi = [data.iloc[:,2].values, data.iloc[:,5].values, data.iloc[:,8].values]

labels = [r"$\nu_e$", r"$\overline{\nu}_{e}$", r"$\nu_x$"]
lss = ["-", "--", ":"]
fs = 16

fig, ax = plt.subplots(1,3, figsize = (16,4))

for j in range(3):
    if j == 0:
        item = luminosity
    elif j == 1:
        item = mean_energy
    elif j == 2:
        item = mean_squared_energy
    for i in range(3):
        if j == 0: ax[j].plot(time, item[i], ls = lss[i], label = labels[i])
        else: ax[j].plot(time, item[i], ls = lss[i])

ax[0].set_xlabel("Time [ms]", fontsize = fs)
ax[0].set_ylabel("Luminosity [erg/s]", fontsize = fs)

ax[0].yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax[0].ticklabel_format(style='sci', axis='y', scilimits=(0,0))

ax[1].set_xlabel("Time [ms]", fontsize = fs)
ax[1].set_ylabel("Mean energy [MeV]", fontsize = fs)

ax[2].set_xlabel("Time [ms]", fontsize = fs)
ax[2].set_ylabel(r"Mean squared energy [MeV$^2$]", fontsize = fs)

for j in range(3): ax[j].tick_params(axis = "both", labelsize = fs)

fig.legend(loc='upper center', ncol=3, bbox_to_anchor=(0.5, 1.1), fontsize = fs)

plt.tight_layout()
```
