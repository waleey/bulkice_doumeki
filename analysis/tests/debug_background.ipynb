```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#from plthelper import *

%config InlineBackend.print_figure_kwargs={'facecolor' : "w"}

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 12
```

```python
filename = "mdom_radioactivity_K40_1s_rand.dat"
output_directory = "../files/output_geant4/"
path = output_directory + "background/" + filename

tmax = 1

data = pd.read_csv(path, sep = "\t", header = None)
data.columns = ["Event ID", "Time [ns]", "Energy [eV]", "PMT ID", "HIT POS X [m]", "HIT POS Y [m]", "HIT POS Z [m]", "VER POS X [m]", "VER POS Y [m]", "VER POS Z [m]", "?", "?", "?"]

print("+++ K40 +++")
print("Simulated number of decays: {:.2f}".format(data.iloc[-1,0]+1))
print("Expected number of decays: {:.2f}".format(61 * 13))
print("Latest time: {:.2f} s".format(data.iloc[:,1].max()/1E9))

plt.plot(data.iloc[:,1]/1E9)
plt.axhline(tmax, c = "red")
plt.xlabel("Entry [#]")
plt.xlim(0, 100)
plt.ylim(0)
plt.ylabel("time stamp [s]")
plt.tight_layout()
plt.savefig("/home/jakob/Desktop/Figure_1.png")
```

```python
filename = "radiodecay_output.csv"
output_directory = "../files/output_geant4/"
path = output_directory + filename

df = pd.read_csv(path, sep = ",")
```

```python
alpha = df.loc[df["ParticleName"] == "alpha"]
electron = df.loc[df["ParticleName"] == "e-"]
photon = df.loc[df["ParticleName"] == "opticalphoton"]
gamma = df.loc[df["ParticleName"] == "gamma"]
```

```python
# Plot alpha spectrum of U238 De
y, bins = np.histogram(alpha["Energy"], bins = 100)
y = 100 * y / alpha.shape[0]
x = (bins[1:]+bins[:-1])/2

fig, ax = plt.subplots(1,1)

ax.step(x,y, where="mid")
ax.set_xlabel("Energy [MeV]")
ax.set_ylabel("Density")
```

```python
# Plot beta spectrum of U238 De
y, bins = np.histogram(electron["Energy"], bins = 100)
y = 100 * y / electron.shape[0]
x = (bins[1:]+bins[:-1])/2 * 1000

fig, ax = plt.subplots(1,1)

ax.step(x, y, where="mid")
ax.set_xlabel("Energy [keV]")
ax.set_ylabel("Density")
ax.set_yscale("log")
```

```python
# Plot photon spectrum of U238 De
y, bins = np.histogram(photon["Energy"], bins = 100)
y = 100 * y / photon.shape[0]
x = (bins[1:]+bins[:-1])/2 * 1000

fig, ax = plt.subplots(1,1)

ax.step(x, y, where="mid")
ax.set_xlabel("Energy [keV]")
ax.set_ylabel("Density")
ax.set_yscale("log")
```

```python
# Plot gamma spectrum of U238 De
y, bins = np.histogram(gamma["Energy"], bins=100)
y = 100 * y / photon.shape[0]
x = (bins[1:]+bins[:-1])/2 * 1000

fig, ax = plt.subplots(1,1)

ax.step(x, y, where="mid")
ax.set_xlabel("Energy [keV]")
#ax.set_xlim(50,100)
ax.set_ylabel("Density")
ax.set_yscale("log")
```
