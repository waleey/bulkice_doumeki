```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#from plthelper import *

%config InlineBackend.print_figure_kwargs={'facecolor' : "w"}

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 12
```

```python
filename = "mdom_radioactivity_K40.dat"
output_directory = "../files/output_geant4/"
path = output_directory + "background/" + filename

data = pd.read_csv(path, sep = "\t", header = None)
data.columns = ["Event ID", "Time [ns]", "Energy [eV]", "PMT ID", "HIT POS X [m]", "HIT POS Y [m]", "HIT POS Z [m]", "VER POS X [m]", "VER POS Y [m]", "VER POS Z [m]", "?", "?", "?"]

print("+++ K40 +++")
print("Simulated number of decays: {:.2f}".format(data.iloc[-1,0]+1))
print("Expected number of decays: {:.2f}".format(61 * 13))
print("Latest time: {:.2f} s".format(data.iloc[:,1].max()/1E9))

plt.plot(data.iloc[:,1]/1E9)
plt.axhline(60, c = "red")
plt.xlabel("Entry [#]")
plt.xlim(0, 100)
plt.ylabel("time stamp [s]")
plt.tight_layout()
plt.savefig("/home/jakob/Desktop/Figure_1.png")
```
