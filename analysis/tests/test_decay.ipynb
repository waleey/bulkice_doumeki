We compare the simulated particle spectra ($\alpha, \beta^{-}, \beta^{+}, \gamma$, optical photons) for each isotope with the IAEA reference data. We load the reference data from the website's API.

We compare simulation generated in the `exp` and `flat` mode. The `exp` mode refers to the mode of sampling decay times from an exponential distribution and verifying whether the daughter would decay in the remaining time. If this mode is activated, the decay products from the daughter decay might be simulated. It is the equivalent to the setting `gRadioSampleExponential=true` in `bulkice_doumeki.cc`. The `flat` mode samples from a flat distribution and only allows subsequent daughter decays for decay times smaller than 10 times the time window. It is the equivalent to the setting `gRadioSampleExponential=false` in `bulkice_doumeki.cc`.

```python
decay_chain = "U238" # "U238", "U235", "Th232"

# directory where geant4 output is saved
output_dir = "~/software/doumeki/bulkice_doumeki/analysis/files/output_geant4/background/mDOM/" + decay_chain 
input_dir = "~/software/doumeki/bulkice_doumeki/mdom/InputFile/" # InputFile directory
input_path = input_dir + decay_chain + "DecayChain_copy.txt"
input_df = pd.read_csv(input_path, delimiter="\t")
isotope_list = input_df["Name"].values

num_particles = 10000
```

```python
mode = "exp"

Isotope = Decay(output_dir, isotope_list, num_particles, mode = mode)
Isotope.run()
```

```python
mode = "flat"

Isotope = Decay(output_dir, isotope_list, num_particles, mode = mode)
Isotope.run()
```

The following shows the hit time of the entire decay chain. We compare again the `exp` with the `flat` mode. We also show the results for the previous method `bad` (this is not meant to be judgemental).

```python
mode = "exp"

Isotope = Decay(output_dir, isotope_list, num_particles, mode = mode)
Isotope.run(flag_comb_hits=True)
```

```python
mode = "flat"

Isotope = Decay(output_dir, isotope_list, num_particles, mode = mode)
Isotope.run(flag_comb_hits=True)
```

```python
mode = "bad"

Isotope = Decay(output_dir, isotope_list, num_particles, mode = mode)
Isotope.run(flag_comb_hits=True)
```

```python
Isotope.
```

```python
fig, ax = plt.subplots(2,1)

ax[0].vlines(Isotope.hit_time_st / 1E9, ymin = 0, ymax = 1, lw = 0.1, alpha = 0.2)
ax[0].axvline(1, color = "k", lw = 2, ls = "--")
ax[0].set_ylim(0,1)
ax[0].set_xlabel("Time [s]")
ax[0].set_yticks([])

ax[1].hist(Isotope.hit_diff_time_st / 1E6, bins = 100, histtype="step", label = "all hits", lw = 2)
ax[1].hist(Isotope.hit_diff_time_st_qe / 1E6, bins = 100, histtype="step", label = "QE biased", lw = 2, ls = "--")

ax[1].set_yscale("log")
ax[1].set_xlabel("Time Difference [ms]")
ax[1].set_ylabel("Counts")
ax[1].legend()

plt.tight_layout()
plt.show()
```
