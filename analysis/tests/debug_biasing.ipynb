```python
import pandas as pd
import numpy as np
from scipy import constants as c

import matplotlib.pyplot as plt
```

```python
dirname = "~/software/doumeki/bulkice_doumeki/analysis/files/output_geant4/signal/mDOM/"

filename_prebias = "prebias_1m.dat"
filename_postbias = "postbias_1m.dat"

path_prebias = dirname+filename_prebias
path_postbias = dirname+filename_postbias

df_prebias = pd.read_csv(path_prebias, delimiter="\t", header = None) # new dataframe
df_prebias.columns = ["event_id", "time", "energy", "pmt_id", "hit_pos_x", "hit_pos_y", "hit_pos_z", 
              "ver_pos_x", "ver_pos_y", "ver_pos_z", "?", "survival", "?"]

df_postbias = pd.read_csv(path_postbias, delimiter="\t", header = None) # new dataframe
df_postbias.columns = ["event_id", "time", "energy", "pmt_id", "hit_pos_x", "hit_pos_y", "hit_pos_z", 
              "ver_pos_x", "ver_pos_y", "ver_pos_z", "?", "survival", "?"]

df_postbias = df_postbias[df_postbias.survival == 1]

energy_min = np.min(np.concatenate([df_prebias.energy.values, df_postbias.energy.values]))
energy_max = np.max(np.concatenate([df_prebias.energy.values, df_postbias.energy.values]))


def get_wavelength(energy): # returns wavelength in nm, input energy in eV
    return (c.Planck * c.speed_of_light) / (energy * c.electron_volt) * 1E9 # wvl = (hc/E) * (1E9 for nm)


wvl_max, wvl_min = get_wavelength(energy_min), get_wavelength(energy_max)
wvl_prebias = get_wavelength(df_prebias.energy.values)
wvl_postbias = get_wavelength(df_postbias.energy.values)
wvl_range = [wvl_min, wvl_max]

fig, ax = plt.subplots(1,1)

ax.hist(wvl_prebias, range=wvl_range, bins = 20, histtype="step", lw = 2, label = "pre-bias")
ax.hist(wvl_postbias, range=wvl_range, bins = 20, histtype="step", lw = 2, label = "post-bias")

ax.set_xlabel("Wavelength [nm]")
ax.set_ylabel("Events")
ax.legend()
```
